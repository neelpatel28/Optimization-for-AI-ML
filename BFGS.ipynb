# BFGS METHOD

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
```

```python
# Objective function
def f_nm(x):
    x1, x2 = x
    return - (x1 - 1)**2 + (x2 - 2)**4
```

```python
def easom(x):
    return np.cos(x[0]) * np.exp(-(x[0] - np.pi)**2)

initial_points = [-5, 5]
results = {}
```

```python
for x0 in initial_points:
    res = minimize(easom, x0=np.array([x0]), method='BFGS')  
    results[x0] = res
    
global_min_x = results[-5].x[0] if results[-5].fun < results[5].fun else results[5].x[0]
global_min_val = results[-5].fun if results[-5].fun < results[5].fun else results[5].fun

print("Global minimum found at:", global_min_x)
print("Function value:", global_min_val)
```

```python
# Plot Easom's function
x_vals = np.linspace(-5, 5, 500)
y_vals = [easom([x]) for x in x_vals]

plt.figure(figsize=(8,5))
plt.plot(x_vals, y_vals, label=r"$f(x) = \cos(x) e^{-(x-\pi)^2}$")
plt.scatter(global_min_x, global_min_val, color="red", label=f"Global Min ({global_min_x:.4f}, {global_min_val:.4f})")
plt.axvline(np.pi, color="gray", linestyle="--", alpha=0.7, label="x = Ï€")
plt.title("Easom's Function and Global Minimum (BFGS)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
